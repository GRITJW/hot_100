```python
class Solution:
    def uniquePaths(self, m: int, n: int) -> int:
        # dp[j]  j 
        dp = [1] * n

        for i in range(1, m):
            for j in range(1, n):
                #  j  =  j dp[j] +  j-1 dp[j-1]
                dp[j] += dp[j-1]
                
        return dp[-1]
m, n = 3, 4
print(Solution().uniquePaths(m, n))
```

O(mn)

O(n)
